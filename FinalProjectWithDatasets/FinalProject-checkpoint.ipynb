```python
X_train, X_test, y_train, y_test = train_test_split( X_reshape, y, train_size= 0.80, test_size = 0.20, random_state = 42)
```

```python
print ('Train data Shape (X,y): ', X_train.shape, ',', y_train.shape)
print ('Test data Shape (X,y): ', X_test.shape, ',',  y_test.shape)
```

```python
#reshape the data to send it through the network
X_train_netShape = X_train.reshape(-1, 100, 100, 1)
X_test_netShape = X_test.reshape(-1, 100, 100, 1)
print (X_train.shape)
print (X_test.shape)
```

```python
#normalize the grayscale values
X_train_netShape_reg = X_train_netShape/255
X_test_netShape_reg = X_test_netShape/255
```

```python
train_Y_one_hot = to_categorical(y_train)
test_Y_one_hot = to_categorical(y_test)
```

```python
print('Train', X_train_netShape_reg[0,0:10,0:10,:], '\n', 'Test', X_test_netShape_reg[0,0:10,0:10,:])
```

```python
train_X,valid_X,train_label,valid_label = train_test_split(X_train_netShape_reg, train_Y_one_hot, test_size=0.2, random_state=42)
```

```python
train_X.shape,valid_X.shape,train_label.shape,valid_label.shape
```

```python
#network structure imports
import keras
from keras.models import Sequential,Input,Model
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras.layers.normalization import BatchNormalization
from keras.layers.advanced_activations import LeakyReLU
```

```python
# batches to start
batch_size = 16
#epochs
epochs = 10
#total class count (binary)
num_classes=2
```

## Final Network Structure

<br>Structure:
<br>Input -> Convolution -> Activation -> Pooling (feature map reduction) -> Convolution -> Activation -> Pooling (feature map reduction) -> Convolution -> Activation -> Pooling (feature map reduction) ->  Flatten image -> Dense Neuron Layer -> Activation -> Output

```python
#Changing the padding to 'valid' and the loss function to a binary cross entropy, I converged to an acceptable model a whole lot faster

gender_model_lin16 = Sequential()
gender_model_lin16.add(Conv2D(32, kernel_size=(3, 3),activation='linear',input_shape=(100,100,1),padding='valid'))
gender_model_lin16.add(LeakyReLU(alpha=0.1))
gender_model_lin16.add(MaxPooling2D((2, 2),padding='valid'))

gender_model_lin16.add(Conv2D(64, (3, 3), activation='linear',padding='valid'))
gender_model_lin16.add(LeakyReLU(alpha=0.1))
gender_model_lin16.add(MaxPooling2D(pool_size=(2, 2),padding='valid'))

gender_model_lin16.add(Conv2D(128, (3, 3), activation='linear',padding='valid'))
gender_model_lin16.add(LeakyReLU(alpha=0.1))                  
gender_model_lin16.add(MaxPooling2D(pool_size=(2, 2),padding='valid'))

gender_model_lin16.add(Flatten())
gender_model_lin16.add(LeakyReLU(alpha=0.1))
gender_model_lin16.add(Dense(num_classes, activation='softmax'))
gender_model_lin16.compile(loss=keras.losses.binary_crossentropy, optimizer='sgd',metrics=['accuracy'])
gender_model_lin16.summary()
gender_training_16 = gender_model_lin16.fit(train_X, train_label, batch_size=batch_size,epochs=epochs,verbose=1,validation_data=(valid_X, valid_label))
gender_model_lin16.save("Best_Gender_Model.hp5y")
```

```python
accuracy = gender_training_16.history['acc']
val_accuracy = gender_training_16.history['val_acc']
loss = gender_training_16.history['loss']
val_loss = gender_training_16.history['val_loss']
epochs = range(len(accuracy))
plt.plot(epochs, accuracy, 'bo', label='Training accuracy')
plt.plot(epochs, val_accuracy, 'b', label='Validation accuracy')
plt.title('Training and validation accuracy')
plt.legend()
plt.figure()
plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()
plt.show()

test_eval = gender_model_lin16.evaluate(X_test_netShape_reg, test_Y_one_hot, verbose=1)
print("The test loss is: %f and the testing accuracy is: %f " % (test_eval[0],test_eval[1]))


predictions = gender_model_lin16.predict(X_test_netShape_reg)
predicted_classes = np.argmax(np.round(predictions),axis=1)
predicted_classes.shape, y_test.shape
correct = np.where(predicted_classes==y_test)
incorrect = np.where(predicted_classes!=y_test)
print("Correct guesses %d \n Incorrect guesses %d"% (len(correct[0]), len(incorrect[0])))
```
