```python
from skimage.transform import resize
import glob
import numpy as np
import matplotlib.image as mpimg
import pandas as pd
```

<h3>Load the Images' Pixels Into an Array</h3>

```python
csv_images = []

for filename in glob.glob('test_square.jpg'):
    color_img = mpimg.imread(filename)
    gray_img = np.dot(color_img[..., :3], [0.2989, 0.5870, 0.1140])
    resized_img = resize(gray_img, (100, 100), anti_aliasing=False)
    resized_img = resized_img.flatten()
    csv_images.append(resized_img)
```

<h3>Convert Array to Numpy Array, then to DataFrame, then to CSV</h3>

```python
csv_images_np = np.array(csv_images)
```

```python
csv_images_np = csv_images_np.astype(int)
```

```python
pd.DataFrame(csv_images_np).to_csv("./images.csv")
```
